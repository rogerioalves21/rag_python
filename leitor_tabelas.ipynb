```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from pymongo import MongoClient
from langchain_mongodb import MongoDBAtlasVectorSearch
from langchain_ollama import ChatOllama, OllamaEmbeddings
from langchain_core.documents import Document
from langchain.chains import RetrievalQA
from langchain.globals import set_debug

set_debug(False)

textos = [("CCI 1081", cci081), ("CCI 500", cci500 ), ("CCI 043", cci043 )]
documents = []
for nome, dados in textos:
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=120, separators=['\x0c'])
    texts = text_splitter.split_text(dados.strip())
    documents.extend([Document(page_content=t, metadata={"source": f"{nome}.PDF", "CCI": nome}) for t in texts])
documents
```
